**Project 2 - Flood Regression**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df = pd.read_csv('flood.csv')
```

```python
df.isnull().sum()
```

```python
X = df.drop(columns=['FloodProbability'],axis=1)
Y = df[['FloodProbability']]
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
Y_predicted = model.predict(X_test)
```

```python
r2 = r2_score(Y_predicted,Y_test)
mse = mean_squared_error(Y_predicted,Y_test)
print(r2)
print(mse)
```

```python
new = [[6,7,4,5,5,5,4,8,8,4,6,8,4,5,4,7,7,5,4,8]]
result=model.predict(new)
print("Flood Probability for the given data :",result)
```
