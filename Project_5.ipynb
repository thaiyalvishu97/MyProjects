Project 5 - Longevity Regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv('Updated Quality of Life Data.csv')
```

```python
df.isnull().sum()
```

```python
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
df['occupation_type'] = le.fit_transform(df['occupation_type'])
df
```

```python
X = df.drop(columns=['id','age_at_death'])
Y = df[['age_at_death']]
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model = DecisionTreeRegressor()
model.fit(X_train,Y_train)
Y_predicted = model.predict(X_test)
```

```python
r2 = r2_score(Y_predicted,Y_test)
mse = mean_squared_error(Y_predicted,Y_test)
print(r2)
print(mse)
```

```python
new = [[0,5,7.73,7.66,7.61,1.00]]
result = model.predict(new)
print("Age of death for given data :",result)
```
