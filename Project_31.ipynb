Project 31 - obesity classification

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("Obesity Classification.csv")
```

```python
df.isnull().sum()
```

```python
df
```

```python
le = LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Label']=le.fit_transform(df['Label'])
df
```

```python
X = df.drop(columns='Label',axis=1)
Y = df['Label']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
model= LogisticRegression()
model.fit(X_train,Y_train)
X_train_predict = model.predict(X_train)
print("Accuracy on training data:",accuracy_score(X_train_predict,Y_train))
X_test_predict = model.predict(X_test)
print("Accuracy on testing data:",accuracy_score(X_test_predict,Y_test))
```

```python
input_data = (4,40,0,150,50,20.0)
input_data_as_num_array = np.asarray(input_data)
input_data_array_reshaped = input_data_as_num_array.reshape(1,-1)
prediction = model.predict(input_data_array_reshaped)
if prediction[0] == 0:
  print("Normal Weight")
elif prediction[0] == 1:
  print("Underweight")
elif prediction[0] == 2:
  print("Overweight")
elif prediction[0] == 3:
  print("Underweight")
```
