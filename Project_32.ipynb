Project 32 - Gender classification

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('gender_classification_v7.csv')
```

```python
df.isnull().sum()
```

```python
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

```python
df
```

```python
X = df.drop(columns='gender',axis=1)
Y = df['gender']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
X_train_predict = model.predict(X_train)
print(X_train_predict)
```

```python
accuracy_X = accuracy_score(X_train_predict,Y_train)
print("Accuracy on training data:",accuracy_X)

X_test_predict = model.predict(X_test)
print("Accuracy on test data:",accuracy_score(X_test_predict,Y_test))
```

```python
input_data = (1,11.8,6.1,1,0,1,1)
input_data_as_num_array = np.asarray(input_data)
input_data_array_reshaped = input_data_as_num_array.reshape(1,-1)
prediction = model.predict(input_data_array_reshaped)

if prediction == 1:
  print("male")
else:
  print("female")
```
