Project 22 - Diamond price prediction regression

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor

df = pd.read_csv("diamonds.csv")
df
df.isnull().sum()
```

```python
df['price'].fillna(df['price'].mean(),inplace=True)
df['x'].fillna(df['x'].mean(),inplace=True)
df['y'].fillna(df['y'].mean(),inplace=True)
df['z'].fillna(df['z'].mean(),inplace=True)
df
```

```python
le = LabelEncoder()
df["cut"] = le.fit_transform(df["cut"])
df["color"] = le.fit_transform(df["color"])
df["clarity"] = le.fit_transform(df["clarity"])
df.head(50)
```

```python
X = df.drop(columns=['price','Unnamed: 0'],axis =1)
Y = df['price']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model = DecisionTreeRegressor()
model.fit(X_train,Y_train)
Y_predict = model.predict(X_test)
r2 = r2_score(Y_predict,Y_test)
mse = mean_squared_error(Y_predict,Y_test)
print(r2)
print(mse)
```

```python
new = [[0.32,1,4,3,63.8,56.0,4.36,4.38,2.79]]
prediction = model.predict(new)
print("predicted Diamond price for the given data is :",prediction[0])
```
