Project 1 - Energy consumption regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('test_energy_data.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
le = LabelEncoder()
```

```python
df['Building Type']= le.fit_transform(df['Building Type'])
df['Day of Week']= le.fit_transform(df['Day of Week'])
df
```

```python
X = df.drop(columns='Energy Consumption',axis=1)
Y = df[['Energy Consumption']]
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
Y_predict = model.predict(X_test)
```

```python
r2 = r2_score(Y_predict,Y_test)
mse = mean_squared_error(Y_predict,Y_test)
print(r2)
print(mse)
```

```python
new = [[2,37943,50,23,21.73,1]]
result = model.predict(new)
print("Energy consumption for given data is :",result)
```
